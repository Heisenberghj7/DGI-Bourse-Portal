### Cleaning The Sides Details

```python
def list_key_value(lst):
    my_dict = {lst[i]: lst[i + 1] for i in range(0, len(lst), 2)}
    return json.dumps(my_dict)
```

```python
sides_details['management'] = sides_details['management'].apply(list_key_value)
```

```python
def pourcentage(lst):
    l=[]
    for value in lst:
        if value.strip() != '%':
            l.append(value)
    return l
```

```python
sides_details["Pourcentage"]=sides_details["Pourcentage"].apply(pourcentage)
```

```python
def unify_paragraph(l):
    par=''
    for p in l:
        par=par+ p.strip()
    return par
```

```python
sides_details["paragraph"]=sides_details["paragraph"].apply(unify_paragraph)
```

```python
def Append_paragraph(l1,l2):
    l2.append(l1)
    return l2
```

```python
sides_details["InfoValue+Paragraph"]=sides_details.apply(lambda row : Append_paragraph(row['paragraph'],row['Info_value']),axis=1)
sides_details["Info_value"] = sides_details["InfoValue+Paragraph"]
sides_details.drop(columns=['InfoValue+Paragraph','paragraph'],inplace=True)
```

```python
def twoCol_key_value(lst1,lst2):
    my_dict = {lst1[i]: lst2[i] for i in range(0, len(lst1))}
    return json.dumps(my_dict,ensure_ascii=False)
```

```python
sides_details['shareholders_pourcentage1'] = sides_details.apply(lambda row: twoCol_key_value(row['shareholders'], row['Pourcentage']), axis=1)
sides_details["shareholders"]=sides_details['shareholders_pourcentage1']
sides_details.rename(columns={"shareholders":"shareholders_pourcentage"},inplace=True)
sides_details.drop(columns=['shareholders_pourcentage1','Pourcentage'],inplace=True)
```

```python
sides_details['info_key_value'] = sides_details.apply(lambda row: twoCol_key_value(row['Info_key'], row['Info_value']), axis=1)
sides_details["Info_key"]=sides_details['info_key_value']
sides_details.drop(columns=['info_key_value','Info_value'],inplace=True)
sides_details.rename(columns={"Info_key":"Info_key_value"},inplace=True)
```

```python
def dividende(l1,l2,l3,l4,l5):
    return [l1,l2,l3,l4,l5]
```

```python
def crossing(l1,l2,l3,l4,l5,l6,l7):
    return [l1,l2,l3,l4,l5,l6,l7]
```

```python
sides_details['dividende'] = sides_details.apply(lambda row: dividende(row['Dividend_years'], row['Dividend_amount'], row['Dividend_type'], row['ex_date'], row['Payment_date']), axis=1)
sides_details['Dividend_years']=sides_details['dividende']
sides_details.rename(columns={"Dividend_years":"dividende_details"},inplace=True)
sides_details.drop(columns=['dividende','Dividend_amount','Dividend_type','ex_date','Payment_date'],inplace=True)
```

```python
sides_details['Crossing_details'] = sides_details.apply(lambda row: crossing(row['crossing_Date'], row['declarant'], row['Quantity'], row['Price'], row['Treshold_crossed'],row['Direction'],row['Intention_of_the_declarant']), axis=1)
sides_details['crossing_Date']=sides_details['Crossing_details']
sides_details.rename(columns={"crossing_Date":"Threshhold crossing"},inplace=True)
sides_details.drop(columns=['declarant','Quantity','Price','Treshold_crossed','Direction','Intention_of_the_declarant','Crossing_details'],inplace=True)
```

```python
sides_details['investors_keys_values'] = sides_details.apply(lambda row: twoCol_key_value(row['investors_keys'], row['investors_values']), axis=1)
sides_details["investors_keys"]=sides_details['investors_keys_values']
sides_details.rename(columns={"investors_keys":"investors_contact"},inplace=True)
sides_details.drop(columns=['investors_keys_values','investors_values'],inplace=True)
```

```python
def fourcolumn_key_value(lst1,lst2,lst3,lst4):
    if len(lst1)== len(lst2)==len(lst3)==len(lst4):
        my_dict = {lst1[i]: [lst2[i],lst3[i],lst4[i]] for i in range(len(lst1))}
        return json.dumps(my_dict)
    return [lst1,lst2,lst3,lst4]
# Years N-1, N-2, N-3
```

```python
sides_details['ratios_values'] = sides_details.apply(lambda row: fourcolumn_key_value(row['ratios'], row['y2022'],row['y2021'],row['y2020']), axis=1)
sides_details['ratios']=sides_details['ratios_values']
sides_details.drop(columns=['ratios_values','y2022','y2021','y2020'],inplace=True)
```

```python
sides_details['figures'] = sides_details.apply(lambda row: fourcolumn_key_value(row['key_figures'], row['2022'],row['2021'],row['2020']), axis=1)
sides_details['key_figures']=sides_details['figures']
sides_details.drop(columns=['figures','2022','2021','2020'],inplace=True)
```

```python
sides_details.drop(columns=['text'],inplace=True)
sides_details.drop(columns=['date'],inplace=True)
```
