```python
import pandas as pd

# Create an empty DataFrame to store the result
Company = pd.DataFrame()

# Iterate through rows of the original DataFrame
for index, row in df.iterrows():
    # Convert the value column to a dictionary (assuming it's a valid JSON-like format)
    values_dict = eval(row['Info_key_value'])
    
    # Create a temporary DataFrame from the dictionary
    temp_df = pd.DataFrame.from_dict(values_dict, orient='index').T
    
    # Append the temporary DataFrame to the result DataFrame
    Company = Company.append(temp_df, ignore_index=True)

# Reset the index of the new DataFrame
Company.reset_index(drop=True, inplace=True)

# Rename columns if needed
Company.rename(columns=lambda x: x.strip(), inplace=True)

#Append the instrument_link, Publications_page columns
Company["Publications_page"]= df["Publications_page"]
Company["instrument_link"]= df["instrument_link"]

# order columns
wanted_order=["Company name","Corporate address","External auditors","Length of fiscal year","Date of creation","Date of IPO","Social object","Publications_page","instrument_link"]
Company=Company[wanted_order]
Company["Company name"]=df["side_full_name"]
Company.rename(columns={'Company name': 'name', 'Corporate address': 'address','External auditors':'external_auditor','Length of fiscal year':'fiscal_year_length','Date of creation':'creation_date','Date of IPO':'ipo_date','Social object':'social_object','Publications_page':'publications_page'}, inplace=True)
Company.creation_date=pd.to_datetime(Company.creation_date)
Company.ipo_date=pd.to_datetime(Company.ipo_date)
# Display the new DataFrame
Company.head()
```
