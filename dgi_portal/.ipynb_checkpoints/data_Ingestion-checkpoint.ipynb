### Spliting the data into tables 

```python
# Create an empty DataFrame to store the result
Company = pd.DataFrame()

# Iterate through rows of the original DataFrame
for index, row in df.iterrows():
    # Convert the value column to a dictionary (assuming it's a valid JSON-like format)
    values_dict = eval(row['Info_key_value'])
    
    # Create a temporary DataFrame from the dictionary
    temp_df = pd.DataFrame.from_dict(values_dict, orient='index').T
    
    # Append the temporary DataFrame to the result DataFrame
    Company = Company.append(temp_df, ignore_index=True)

# Reset the index of the new DataFrame
Company.reset_index(drop=True, inplace=True)

# Rename columns if needed
Company.rename(columns=lambda x: x.strip(), inplace=True)

# Append the instrument_link, Publications_page columns
Company["Publications_page"]= df["Publications_page"]
Company["instrument_link"]= df["instrument_link"]

# order columns
wanted_order=["Company name","Corporate address","External auditors","Length of fiscal year","Date of creation","Date of IPO","Social object","Publications_page","instrument_link"]
Company=Company[wanted_order]
Company["Company name"]=df["side_full_name"]
Company.rename(columns={'Company name': 'name', 'Corporate address': 'address','External auditors':'external_auditor','Length of fiscal year':'fiscal_year_length','Date of creation':'creation_date','Date of IPO':'ipo_date','Social object':'social_object','Publications_page':'publications_page'}, inplace=True)
Company.creation_date=pd.to_datetime(Company.creation_date)
Company.ipo_date=pd.to_datetime(Company.ipo_date)
# Display the new DataFrame
Company.head()
```

```python
engine= create_engine("postgresql://root:root@localhost:5432/bourse")
```

```python
engine.connect()
```

```python
Company.to_sql(name='api_company',index=False,con=engine,if_exists='append')
```

```python
key_list = []
value_list = []
sides=[]


management_list = df['management'].tolist()
side_full_name_list = df['side_full_name'].tolist()

for management_dict, side_full_name in zip(management_list, side_full_name_list):
    for key, value in eval(management_dict).items():
        key_list.append(key)
        value_list.append(value)
    mult_sides= [side_full_name]*len(eval(management_dict))
    sides.extend(mult_sides)         
Management = pd.DataFrame({
    'post': key_list,
    'name': value_list,
    'company_id': sides
})
wanted_order=['name','post','company_id']
Management=Management[wanted_order]
```

```python
Management.head()
```

```python
Management.to_sql(name='api_management',index=False,con=engine,if_exists='append')
```

```python
key_list = []
value_list = []
sides=[]


shareholders_list = df['shareholders_pourcentage'].tolist()
side_full_name_list = df['side_full_name'].tolist()

for shareholder_dict, side_full_name in zip(shareholders_list, side_full_name_list):
    for key, value in eval(shareholder_dict).items():
        if key!='Total':
            key_list.append(key)
            value_list.append(value)
    mult_sides= [side_full_name]*(len(eval(shareholder_dict))-1)
    sides.extend(mult_sides)  
shareholders = pd.DataFrame({
    'holder': key_list,
    'percentage': value_list,
    'company_id': sides
})
```

```python
shareholders.head()
```

```python
shareholders.to_sql(name='api_shareholder',index=False,con=engine,if_exists='append')
```

```python
# Create an empty DataFrame to store the result
Dividend = pd.DataFrame()
side_full_name_list = df['side_full_name'].tolist()
index=0
# Iterate through rows of the original DataFrame
for index, row in df.iterrows():
    details = row['dividende_details']
    dt= list(map(list, zip(*details)))
    # Create a temporary DataFrame from the dictionary
    temp_df = pd.DataFrame(dt, columns=['year', 'amount', 'dividend_type','ex_date','payment_date'])
    temp_df['company_id']= [side_full_name_list[index]]*len(details[0])
    index+=1
    # Append the temporary DataFrame to the result DataFrame
    Dividend = Dividend.append(temp_df, ignore_index=True)
        
# Reset the index of the new DataFrame
Dividend.reset_index(drop=True, inplace=True)

# Rename columns if needed
Dividend.rename(columns=lambda x: x.strip(), inplace=True)
Dividend.payment_date=pd.to_datetime(Dividend.payment_date)
Dividend.ex_date=pd.to_datetime(Dividend.ex_date)

# Display the new DataFrame
Dividend.head()
```

```python
import pandas as pd

KeyFigures = pd.DataFrame()
side_full_name_list = df['side_full_name'].tolist()

# Iterate through rows of the original DataFrame
for index, row in df.iterrows():
    
    # Convert the value column to a dictionary (assuming it's a valid JSON-like format)
    values_dict = eval(str(row['key_figures']))  # Assuming this is a string representation of a dictionary
    
    # Check if values_dict is a dictionary before proceeding
    if isinstance(values_dict, dict):
        values_dict['year']=['N-1','N-2','N-3']
        for key in values_dict.keys():
            values_dict[key] = list(values_dict[key])
        # Create a temporary DataFrame from the dictionary
        temp_df = pd.DataFrame(values_dict)
        temp_df['company_id']=side_full_name_list[index]
        # Append the temporary DataFrame to the result DataFrame
        KeyFigures = KeyFigures.append(temp_df, ignore_index=True)

# Reset the index of the new DataFrame
KeyFigures.reset_index(drop=True, inplace=True)

# Rename columns if needed
KeyFigures.rename(columns=lambda x: x.strip(), inplace=True)

interm=KeyFigures['Consolidated accounts'].copy()
KeyFigures['Consolidated accounts']=KeyFigures['year']
KeyFigures['year']=interm

KeyFigures.rename(columns={'Consolidated accounts': 'year','Share capital':'share_capital','Equity':'equity','Number of titles(2)':'number_of_titles','Sales figures':'sales_figures','Operating result':'operating_result','Net income (4)':'net_income','year':'consolidated_accounts'}, inplace=True)




KeyFigures.head()
```

```python
KeyFigures.to_sql(name='api_keyfigures',index=False,con=engine,if_exists='append')
```

```python
import pandas as pd

Ratios = pd.DataFrame()
side_full_name_list = df['side_full_name'].tolist()

# Iterate through rows of the original DataFrame
for index, row in df.iterrows():
    
    # Convert the value column to a dictionary (assuming it's a valid JSON-like format)
    values_dict = eval(str(row['ratios']))  # Assuming this is a string representation of a dictionary
    
    # Check if values_dict is a dictionary before proceeding
    if isinstance(values_dict, dict):
        values_dict['year']=['N-1','N-2','N-3']
        for key in values_dict.keys():
            values_dict[key] = list(values_dict[key])
        # Create a temporary DataFrame from the dictionary
        temp_df = pd.DataFrame(values_dict)
        temp_df['company_id']=side_full_name_list[index]
        # Append the temporary DataFrame to the result DataFrame
        Ratios = Ratios.append(temp_df, ignore_index=True)

# Reset the index of the new DataFrame
Ratios.reset_index(drop=True, inplace=True)

# Rename columns if needed
Ratios.rename(columns=lambda x: x.strip(), inplace=True)

Ratios.insert(0,'yearr',Ratios.year)
Ratios.drop('year',inplace=True,axis=1)
Ratios.rename(columns={'yearr': 'year','BPA':'bpa','ROE (in %)':'roe','Payout (in %)':'payout','Dividend yield (in %)':'dividend_yield','PER':'per','PBR':'pbr'}, inplace=True)


Ratios.head()
```

```python
Ratios.to_sql(name='api_ratios',index=False,con=engine,if_exists='append')
```
